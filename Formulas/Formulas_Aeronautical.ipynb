# Basics of Thermodynamics

>## Equation of state
  $$ \frac{P}{ρ.T} = Constant = R$$
 <br>
  $$ P.v = R.T $$
  <br>
  $$ \frac{P}{ρ} = R.T$$
  <br>
  $$P.V = m.R.T$$
  <br>
  Properties of State
  - $P$ - Pressure
  - $v$ - Specific Volume
  - $ρ$ - Density
  - $T$ - Temperature
  - $R$ - Characterstic Gas constant
  - $m$ - Molar mass of Gas


> ## Characteristic Gas Constant ($R$)
  $$R = \frac{R_{o}}{M}$$
  - $R$ = Characteristic Gas Constant
  - $R_{o}$ = Universal Gas Constant = $8314 \, \left(\frac{J}{Kg \cdot K}\right)$
  - $M$ = Molar mass of the gas\
  <br>
  Characteristic Gas constant for gases:
    - $R_{O_{2}} = \frac{R_{o}}{M_{O_{2}}} = \frac{R_{o}}{32} = 259.8 \, \left(\frac{J}{Kg \cdot K}\right)$
    - $R_{N_{2}} = \frac{R_{o}}{M_{N_{2}}} = \frac{R_{o}}{28} = 296.92 \, \left(\frac{J}{Kg \cdot K}\right)$
    - $R_{Air} = \frac{R_{o}}{M_{Air}} = \frac{R_{o}}{28.96} = 287 \, \left(\frac{J}{Kg \cdot K}\right)$


> ## Relationship between properties in different Processes
  **Constant Volume Process (Gay-Lussac's Law)**
    $$\frac{P_{2}}{P_{1}} = \frac{T_{2}}{T_{1}}$$
  **Constant Pressure Process (Charles's Law)**
    $$\frac{V_{2}}{V_{1}} = \frac{T_{2}}{T_{1}}$$
  **Constant Temperature Process (Boyle's Law)**
    $$ P_{1}.V_{1} = P_{2}.V_{2} $$




> ## Displacement work done
  $$ W = ∫P.dv $$
  Work done = Area under $(P-V)$ Curve
  - Units
    - Work done  = $ Joule $
    - Workdone for unit mass =  $\frac{Joule}{Kg}$
  - **Work done in differnt Processes**
    - Constant volume $(V_{1} = V_{2}) , dv =0$
      $$ W = ∫P.dv = 0 $$
    - Constant Pressure $(P_{1} = P_{2})$
      $$ W = \int_{1}^{2} P.dv = P(V_{2} - V_{1})$$
    - Constant Temperature $(T_{1} = T_{2})$
      $$ W = ∫P.dv = P_{1}.V_{1}.\ln( \frac{V_2}{V_1} )$$  
      $$ W = ∫P.dv = P_{1}.V_{1}.\ln( \frac{P_1}{P_2} )$$  
    - Isentropic Process $( ΔS = 0 )$ ${P.v^{γ} = C}$
      $$ W = \frac{P_1.V_1 - P_2.V_1}{ γ - 1} $$
    - Polytropic Process $(P.v^n = C)$
      $$ W = \frac{P_1.V_1 - P_2.V_1}{ n - 1} $$

> ## Slope of polytropic equation
  $$ \frac{dP}{dv} = \frac{-np}{V} $$
  - n - Polytropic index

> ## Specific heat of gas
  - At constant volume $(C_v)$
  $$ dU = C_v.dT $$
  - At constant Pressure $(C_p)$
  $$ dh = C_p.dT $$

> ## Adiabatic index $(γ)$
  $$ γ = \frac{C_p}{C_v} $$
  - Range for γ
    - for gases $γ > 1$
    - range of $γ$ is $ 1 \leq γ \leq 1.67 $
  - γ for gases
    - for Monoatomic gases $(He)$ ---------------  $ γ = 1.67 $
    - for Diatomic gases $(O_2,N_2,Air)$ -------  $ γ = 1.4  $
    - for Triatomic gases $ (CO_2) $ -------------------  $ γ = 1.28 $
  - Molecular-Mass (⇑) ------  γ (⇓)


> ## Relationship between $C_p, C_v, R, γ$
  $$ C_p - C_v = R $$
  $$ C_P = \frac{γ.R}{γ-1} $$
  $$ C_v = \frac{R}{γ-1} $$
  - For Air
    * $C_p = 1005 \frac{Joule}{Kg - K}$
    * $ C_v = 717 \frac{Joule}{Kg -K}$

> ## Flow Work (Steady Flow Energy Equation)
  $$ h_1 + \frac{{V_1}^2}{2} + g.Z_1 + q =  h_2 + \frac{{V_2}^2}{2} + g.Z_2 + W_s$$
  - Applicaitons
    - Duct (Nozzzle/ Diffuser)
      $$ h_1 + \frac{{V_1}^2}{2} =  h_2 + \frac{{V_2}^2}{2} $$
      $$ C_p.T_1 + \frac{{V_1}^2}{2} =  C_p.T_2 + \frac{{V_2}^2}{2} \ \ Here, V(⇑) T(⇓) $$
    - Turbine / Compressor
      $$ W_s = C_p.(T_1 - T_2) $$


> ## First Law of Thermodynamics
  $$ dQ = dU + dW $$
  *   Q - Heat
  *   U - Internal energy
  *   W - Work done


> ## Entropy $(S)$
  Randomness increases upon heat added
  $$ dS = \frac{dq}{T} $$
  Units = $ (\frac{Joule}{Kg - K})$

> ## $T.dS$ Equations
  from first law
    $$ T.dS = dU + P.dv $$
  from enthaply relation
    $$ dh = T.dS + v.dP $$

> ## Change in Entropy $(ΔS)$
  $$ ΔS = C_p.\ln(\frac{T_2}{T_1})- R.\ln(\frac{P_2}{P_1}) $$
  $$ ΔS = C_v .\ln(\frac{T_2}{T_1} + R.\ln(\frac{V_2}{V_1})) $$

> ## Plot of $T-dS$ Plane
  - Constant Temperature $ ( ΔT = 0) $
  $$ \frac{dT}{dS} = 0 $$
  - Constant Pressure process $( ΔP=0)$
  $$ \frac{dT}{dS} = \frac{T}{C_p} $$
  - Constant Volume process $( ΔS = 0 )$
  $$ \frac{dT}{dS} = \frac{T}{C_v} $$
  - Isentropic process $( ΔV = 0 )$
  $$ \frac{dT}{dS} = ∞$$

> ## Isentropic Equaiton
  $$ P.v^γ = Constant $$
  - Relationships between $P,V,T,ρ$
    - $$ \frac{P_2}{P_1} = \left(\frac{ρ_2}{ρ_1} \right)^γ$$
    - $$ \frac{P_2}{P_1} = \left( \frac{T_2}{T_1} \right)^{\frac{\gamma}{\gamma-1}} $$
    - $$ \frac{ρ_2}{ρ_1} = \left( \frac{T_2}{T_1} \right)^{\frac{1}{\gamma-1}} $$



# Gas Dynamics

> ## Volumetric Strain $(\frac{dV}{V})$
  $$ \frac{dV}{V} = \frac{1}{V} \left (\left (\frac{δV}{δT} \right)dT \right) +
  \frac{1}{V}\left (\left (\frac{δV}{δP} \right)dP \right) $$
  -  $\frac{1}{V}\left (\left (\frac{δV}{δP} \right)dP \right)$ is compression
  - $ \frac{1}{V} \left (\left (\frac{δV}{δT} \right)dT \right) $ is expansion


> ## Compressibilty $ (τ) $
  $$ τ =  - \left (\frac{1}{V} \right ) \left(\frac{dV}{dP} \right) = \left (\frac{1}{ρ} \right ) \left(\frac{dρ}{dP} \right)$$
  - Constant Temperature $(T=C)$
  $$ \left (τ\right)_{Isothermal} = \frac{1}{P} $$
  - Isentropic process $( Δ S = 0)$
  $$ \left (τ\right)_{Isothermal} = \frac{1}{\gamma P} $$

> ## Euler's differerantial equation of motion
  $$ \frac{dP}{\rho} + V.dV = 0$$

> ## Bernoulli's Equaiton
  - For incompressible flow
  $$ P + \left (\frac{1}{2} \right) \rho V^2 = C$$
  $$ P_1 + \left (\frac{1}{2} \right) \rho_1 {V_{1}}^2 = P_2 + \left (\frac{1}{2} \right) \rho_2 {V_{2}}^2$$
  - For compressible flow (isentropic flow)
  $$ \frac{\gamma P}{ρ(γ-1)} + \left (\frac{1}{2} \right)V^2 = C $$
  $$ \frac{\gamma P_1}{ρ_1(γ-1)} + \left (\frac{1}{2} \right){V_{1}}^2 = \frac{\gamma P_2}{ρ_2(γ-1)} + \left (\frac{1}{2} \right){V_{2}}^2  $$

> ## Mass Flow rate $(\dot{m})$
  $$ \dot{m} = ρ.A.V = Constant $$
  - Nomenclature
    - ρ - Density
    - A - Area
    - V - Velocity of the flow
  $$ ρ_1 A_1 V_1 = ρ_2 A_2 V_2 $$
  * For incompressible flow $(ρ=Constant)$
   $$ A_1 V_1 = A_2 V_2$$
  Volumetric flow $(Q)$ = $ AV (\frac{m^3}{s}) $

> ## Speed of sound $(a)$
  $$ a = \sqrt{\frac{dP}{d\rho}} $$
  - For Isentropic flow
  $$ a = \sqrt{\frac{γP}{ρ}} = \sqrt{γRT} $$

> ## Mach Number $(M)$
  $$ M = \frac{Speed of Object}{Local speed of the sound} = \frac{V}{a} $$
  - Mach number and flow regimes
    - $M < 0.3$  -  Incompressible flow
    - $0.3 \leq M \leq 0.8$  - Transonic flow
    - $ M < 1$ - Subsonic flow
    - $ M = 1$ - Sonic flow
    - $ M > 1$ - Supersonic flow
    - $ M > 5$ - Hypersonic flow
  
> ## Concept of Stagnation and Static Properties
  Stagnation properties are represented as $P_o, T_o, h_o, \rho_o \ at \ Velcity =0$ \
  $$ h_o = h + \frac{V^2}{2} $$
  $$ T_o = T + \frac{V^2}{2C_P} $$
  $$ \frac{T_0}{T} = 1 + (\frac{γ-1}{2})M^2 $$
  $$ \frac{P_o}{P} = \left (\frac{T_o}{T} \right)^{\frac{γ}{γ-1}} = \left (1 + (\frac{γ-1}{2})M^2 \right)^{{\frac{γ}{γ-1}}}$$
  $$ \frac{\rho_o}{ρ} = \left (\frac{T_o}{T} \right)^{\frac{1}{γ-1}} = \left (1 + (\frac{γ-1}{2})M^2 \right)^{{\frac{1}{γ-1}}}$$

> ## Area-Velocity relation with respect to Mach number
  $$ \frac{dA}{A} = \frac{dV}{V}(M^2 - 1) $$
  - $ V (⇑) M(⇑)$ then $ P(↓), T(↓), ρ (↓)$
  - $ V (↓) M(↓)$ then $ P(⇑), T(⇑), ρ (⇑)$

> ## Mach Angle
  $$ μ = \sin^{-1}(\frac{1}{M}) $$

> ## Critical properties
  - the properties  of gas when Speed of the flow or object is exactly equal to local speed of sound
  <br>
  - they are $ P^* , T^* , \rho^* , a^* \ at \ M = 1$\
  $$ \frac{T_0}{T} = 1 + (\frac{γ-1}{2})M^2 = 1 + (\frac{γ-1}{2})(1)^2  = \frac{γ+1}{2}$$
  $$ \frac{T_0}{T} = \frac{γ+1}{2}$$
  $$ \frac{P_o}{P} = \left (\frac{γ+1}{2} \right)^{\frac{γ}{γ-1}} $$
  $$ \frac{\rho_o}{ρ} = \left (\frac{γ+1}{2} \right)^{\frac{1}{γ-1}} $$
  -For Air
    - $\frac{T_0}{T} = 1.2$
    - $ \frac{P_o}{P} = 1.89$
    - $ \frac{\rho_o}{ρ} = 1.57 $

> ## Characterstic Mach Number $(M^*)$
  $$ M^* = \frac{Speed \ of \ the \ object}{Critical \ speed \ of \ sound} = \frac{V}{a^*} $$

> ## Energy equation in terms of speed of sound
  $$ h_1 + \frac{{V_1}^2}{2} =  h_2 + \frac{{V_2}^2}{2} $$
  $$ \frac{{a_1}^2}{γ-1} + \frac{{V_1}^2}{2} = \frac{{a_2}^2}{γ-1} + \frac{{V_2}^2}{2} $$
  - ### For stagnation speed of sound
  $$ \frac{{a_0}^2}{γ-1} = \frac{{a}^2}{γ-1} + \frac{{V}^2}{2} $$
  - ### For $M = 1$ then $V=a^*$ { relation with Critical properties }
  $$ \frac{{a}^2}{γ-1} + \frac{{V}^2}{2} = \frac{{a^*}^2}{γ-1} + \frac{{a^*}^2}{2}$$
  $$ \frac{{a}^2}{γ-1} + \frac{{V}^2}{2} = \frac{({γ+1}){a^*}^2}{2(γ-1)} $$

> ## Relation between $M$ and $M^*$
  $$ {M^*}^2 = \frac{(γ+1)M^2}{2+(γ-1)M^2} $$
  if $M → ∞$ then $ M^* = \sqrt{\frac{γ+1}{γ-1}} $
  - for air
    - if $M → ∞$ then $ M^* = \sqrt{6} = 2.45 $



> ### Continuty Equation
  - Integral form of continuty equation
   $$  \frac{\partial }{∂t}\iiint ρ.∂v  + \iint ρ V.∂s =0 $$
    - for Steady Flow
     $$ \iint ρ V.∂s =0 $$
    - Derived form
     $$ ρ_1 A_1 V_1 = ρ_2 A_2 V_2 $$
      - if $A_1 = A_2$
        $$ ρ_1  V_1 = ρ_2 V_2 $$
      - if $\rho = constant$
        $$ A_1 V_1 = A_2 V_2 $$

> ### Momentum Equation
  - Integral form of momentum equation
  $$  \frac{\partial }{∂t}\iiint (ρ.∂v)V  + \iint (ρ V.∂s)V = \iiint \rho \ f \  \partial v  - \iint P.∂s + F_{v}$$
    - $f$ = body force,
    - $F_{v}$ = Viscous force
  - if Flow is steady, No body force and invisid
     $$ \iint (ρ V.∂s)V = \iiint \rho \ f \  \partial v  $$\
     $$ P_1 + ρ_1 {V_1}^2 = P_2 + ρ_2 {V_2}^2 $$

> ### Energy Equation
  $$ C_p . T_1 + \frac{{V_1}^2}{2} = C_p . T_2 + \frac{{V_2}^2}{2} $$




> ## Normal Shock
  $$ {M_2}^*  = \frac{1}{{M_1}^*}$$
  - Mach number after normal shock $M_2$
  $$ {{M_2}^2} = \frac{1 + ( \frac{γ - 1}{2} ) * {{M_1}^2}} { γ {{M_1}^2} - \frac{γ -1}{2} }$$
    - if $ M_1 ⟶ ∞$
      $$ M_2 = \sqrt{\frac{γ-1}{2γ}} $$
    - for Air $ M_1 ⟶ ∞$ then $ M_2 = 0.378$
      - if $M_1 < ∞ $,   then   $M_2 > 0.378$
  - ### Pressure ratio $(\frac{P_2}{P_1})$
     $$ \frac{P_2}{P_1} = 1 + \frac{2γ}{γ+1}({M_1}^2 -1) $$
  - ### Density ratio $(\frac{\rho_2}{\rho_1})$
     $$ \frac{\rho_2}{\rho_1} = \frac{V_1}{V_2} = {{M_1}^*}^2 = \frac{(γ +1){M_1}^2}{ 2 + (γ - 1){M_1}^2 } $$
  - ### Temperature ratio $(\frac{T_2}{T_1})$
     $$ \frac{T_2}{T_1} = \frac{P_2}{P_1} * \frac{\rho_1}{\rho_2} $$  
  - ### Entropy change $(ΔS)$
     $$ ΔS = -R \ln {\frac{P_{02}}{P_{01}}} $$

> ## Max Velocity of flow from reservoir

> ## Moving Normal Shock
  - Earth attached cordinate system
    - Speed of air before shock - $U_1$
    - Speed of air after shock - $U_2$
    - Speed of shock - $C_s$
  - Shock attached cordinatee system
    - Speed of air before shock - $V_1 = C_s - U_1$
    - Speed of air after shock - $V_2 = C_s = U_2
  - Solution approch
    - Find $M_1$ $$ M_1 = \frac{V_1}{a_1} $$
    - by using denisty ratio, find $V_2$ $$ \frac{\rho_2}{\rho_1} = \frac{V_1}{V_2} = {{M_1}^*}^2 = \frac{(γ +1){M_1}^2}{ 2 + (γ - 1){M_1}^2 } $$
    - caluculate $U_2$
    - Find Pressure ratio $$ \frac{P_2}{P_1} = 1 + \frac{2γ}{γ+1}({M_1}^2 -1) $$
    - find Temperature ratio and get $T_2$ $$\frac{T_{2}}{T_{1}} = \frac{P_{2}}{P_{1}} * \frac{ρ_{1}}{\rho _{2}} $$

>## Oblique Shock Wave
  If flow turns toward itself then oblique shock is formed
  - Nomencalature
    - Deflection angle - $θ$
    - Shock angle - $β$
  - Approch to solution
    -  you have $M_1$  find $M_{n1}$   $$ M_{n1} = M_1 . sin{(β)} $$
      * Note to form shock wave $M_{n1} > 1$
        $$ β > {sin}^{-1}(\frac{1}{M_1}) $$
      * $β>μ$ , where $ μ$ is Mach angle
    - Now can use Normal shock formulas for Normal Mach numbers
      * to find $M_{n2}$
        $$ {M_{n2}}^2 = \frac{ 1 + \left( \frac{γ-1}{2} \right) {M_{n1}}^2 }{γ.{M_{n1}}^2 - \left (\frac{γ-1}{2}\right)} $$
      * but you need $M_2$
        $$ M_2 = \frac{M_{n2}}{Sin(β-θ)} $$
      * Pressure ratio
        $$ \frac{P_2}{P_1} = 1 + \frac{2γ}{γ+1}({M_{n1}}^2 -1) $$
      * Density ratio
        $$ \frac{\rho_2}{\rho_1} = \frac{V_{n1}}{V_{n2}} = {{M_{n1}}^*}^2 = \frac{(γ +1){M_{n1}}^2}{ 2 + (γ - 1){M_{n1}}^2 } $$

> ## θ - β - M  Relation
  $$ tan(θ) = \frac{2Cot(β) \left( {M_1}^2.Sin^2(β) - 1 \right)}{ {M_1}^2 \left( γ.Cos(2β) \right) + 2} $$
  Discussion on the relation
  1. if $θ = θ_{Max}$ \
    - There is no solution for straight oblique shock, The shock detaches and forms Bow shock
  2. if Mach $(⇑)$ then $θ_{Max} (⇑)$
  3. if $θ < θ_{Max}$ you get 2 solutions
    * Weak solution
    * Strong solution
  4. if $θ = 0$
    * $β = \frac{π}{2}$ Normal shock
    * $β = μ$ for No shock formation
  5. for $θ - β - M $ relation
    * if $ θ $ = Constant
      - For Weak solution
        * if M (⇑) then β (⇓)
        * if M (⇓) then β (⇑)
      - For Strong solution
        * if M (⇑) then β (⇑)
        * if M (⇓) then β (⇓)
    * if $M$ = Constant
      - For weak solution
        * if θ (⇑) then β (⇑)
        * if θ (⇓) then β (⇓)
      - For strong solution
        * if θ (⇑) then β (⇓)
        * if θ (⇓) then β (⇑)
        
> ## Expansion wave
  This Flow is isentropic and formed when flow turns away from itself
  - **Prandtl - Meyer Function** ( $\bf ν$ )
  $$ \bf{ν}  = \left(\sqrt{\frac{γ+1}{γ-1}} \right) tan^{-1}{\left( \sqrt{\frac{γ-1}{γ+1} (M^2 -1)} \right)} - tan^{-1}(\sqrt{M^2 -1}) $$
    - for Air
      $$ \bf ν  = \sqrt{6} tan^{-1}{\left(  \sqrt{\frac{(M^2 -1)}{6}} \right)} - tan^{-1}(\sqrt{M^2 -1}) $$
  - Solution approch
    1. for given $M_1$ caluculate $\bf ν_1$
    2. find $\bf ν_2$ $$ \bf ν_2 = ν_1 + θ $$
    3. find $M_2$ using $\bf ν_2$
    4. Apply Isentropic relations
  - if $M_1$ = 1 and $M_2$ = ∞
    - if $M_1 = 1$
      - $\bf ν_1 = 0$
    - if $ M_2 = ∞$
      - $ \bf ν_2 = \left [ \sqrt{\frac{γ+1}{γ-1}} -1 \right]\frac{π}{2}$
      - $θ =  θ_{Max} = \bf ν_2$
  $$ θ_{Max} =  \left [ \sqrt{\frac{γ+1}{γ-1}} -1 \right]\frac{π}{2}  $$
  - Fan angle
    $$ fan \ angle = μ_1 - μ_2 + θ $$
